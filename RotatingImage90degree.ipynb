```python
import cv2
%matplotlib inline
import matplotlib.pyplot as plt 


def rotate_image(image,p = 'ok') :
    rows = len(image)
    cols = len(image[0])
    arr = [[0]*rows for i in range(cols)]
    for i in range(cols) :
        for j in range(rows) :
            arr[i][j] += image[j][i]
    if p == 'ok' :
       print(arr)
    else :
       return arr
    
def showImage(im) :
    plt.imshow(im,cmap='gray')
    plt.show()
```
